```python
from zipfile import ZipFile
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
with ZipFile('/content/youtube+spam+collection.zip','r') as zip:
    zip.extractall('./spam')
```

```python
psy_df = pd.read_csv('/content/spam/Youtube01-Psy.csv')
katy_df = pd.read_csv('/content/spam/Youtube02-KatyPerry.csv')
lmfao_df = pd.read_csv('/content/spam/Youtube03-LMFAO.csv')
eminem_df = pd.read_csv('/content/spam/Youtube04-Eminem.csv')
shakira_df = pd.read_csv('/content/spam/Youtube05-Shakira.csv')

spam_df = pd.concat([psy_df,katy_df,lmfao_df,eminem_df,shakira_df],axis=0)
```

```python
spam_df.head()
```

```python
# Data processing del campo CONTENT
spam_df['CONTENT'] = spam_df['CONTENT'].str.lower()
spam_df['CONTENT'] = spam_df['CONTENT'].str.replace('[^\w\s]','')
spam_df['CONTENT'] = spam_df['CONTENT'].str.replace('\d+','')
spam_df['CONTENT'] = spam_df['CONTENT'].str.strip()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Tokenización del campo CONTENT
tfidf_vectorizer = TfidfVectorizer(max_features=1000)

X = tfidf_vectorizer.fit_transform(spam_df['CONTENT'])

print("Shape of TF-IDF matrix:", X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, spam_df['CLASS'], test_size=0.2, random_state=42)
```

```python
bayes_model = MultinomialNB()
bayes_model.fit(X_train,y_train)
```

```python
y_pred = bayes_model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
